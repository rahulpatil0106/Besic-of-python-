# Seaborn 
 
 
 statistical visualization library

## Distribution plot

```python
import seaborn as sns 
```

```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.info()
tips.head()
```

```python
tips["time"].unique()
```

## Distribution plot

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"],kde=False)
```

```python
sns.distplot(tips["total_bill"],kde=True,bins=20)
```

## sns.jointplot()

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')   # 'reg' for regression line
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind="kde'")
```

## sns.pairplot()

plots the jiontplot of all possiable combinations of numeric variables

```python
sns.pairplot(data=tips)
```

```python
sns.pairplot(data=tips,hue='sex')
```

## categorical data plot

## sns.coutplot()

```python
sns.countplot(x='sex',data=tips)
```
