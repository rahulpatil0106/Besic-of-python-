```python
import matplotlib.pyplot as plt 
%matplotlib inline 
import numpy as np 
```

```python
x=[1,2,3,4,5,6,7,8]
y=[5,8,94,6,89,56,47,1]
plt.scatter(x,y)
```

```python
x = np.linspace(0,5,num= 100)
y = x**2
```

### Functional Method for ploting with matplot lib 

```python
plt.plot(x,y)
plt.xlabel('higth')
plt.ylabel('hight**2')
plt.title('Hight Vs Hight Square')
```

```python
##  ploting multiple plots on single graphe 
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(y,x,'b')
plt.subplot(2,2,3)
plt.plot(x ,x**3,'g')
plt.subplot(2,2,4)
plt.plot(x**3,y,'y')
```

### Object orianded ploting with matplotlib


```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.2,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X variable')
axes.set_ylabel('Y variable')
axes.set_title('This is plot')
```

```python
fig = plt.figure()
axes1= fig.add_axes([0.1,0.2,0.8,0.8])
axes1.plot(x,x**3)


axes2 = fig.add_axes([0.15,0.6,0.5,0.3])
axes2.plot(x**3,y,'r')
```

## Subplot using object oriented programimg 


```python
fig ,axes = plt.subplots(nrows = 1,ncols = 2)
axes[0].plot(x,y)

axes [1].plot(y,x)
```

### Fig Size and dip 

```python
#SimpleOO ploting 
fig = plt.figure(figsize=(3,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1, ncols =2 ,figsize=(5,5))
axes[0].plot(y,x)
axes[1].plot(x,y,'r')
```

```python
## Legent and locator in matplotlib

fig  = plt.figure(dpi= 50)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label= 'X**2')
axes.plot(x,x**3,label='X**3')

axes.legend()
```

## plot Appearance 

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,lw =5,ls = '--',color='red')    # lw : for line width  ,ls : for line style ,color:  color of line 
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
z = np.linspace(0,5,10)
axes.plot(z,z**2,lw=1,ls='-',marker='o',color ='red',markerfacecolor= 'blue',markersize =20,markeredgecolor='black',markeredgewidth = 2)
```
